# **DDC-4: Bar plot of word lengths**
Create a string variable for the sentence “This is a sentence that should say something witty but instead is just meta.”

Split the sentence into words, count the number of characters in each word, and show the results in a bar graph as below.
<p align="center">
  <img src="images/img_ddc_04.png" alt="04" width="700">
</p>

```python
# Libs
import matplotlib.pyplot as plt
import matplotlib as mpl
import re

plt.style.use("dark_background")
```

```python
text = "This is a sentence that should say something witty but instead is just meta."
words = re.split(" ",text)
words
```

```python
# ?plt.bar
```

```python
plt.figure(figsize=(11,4))

bars = plt.bar(range(len(words)),[len(w) for w in words])

for idx in range(len(words)):
    bars[idx].set_color(mpl.cm.plasma((1+idx)/len(words)))
    
plt.xticks(rotation=45,ticks=range(len(words)),labels=words)
plt.gca().set(ylabel="character counts")
plt.tight_layout()
plt.show()
```
