```python
# Libs

import matplotlib.pyplot as plt
import numpy as np

plt.style.use("dark_background")
```

```python
x   = np.linspace(0.001,3,131)
ex  = np.exp(x)
lny = np.log(x)
```

```python
_,ax = plt.subplots(1)

ax.plot(x,ex,"g--",label="$e^x$")
ax.set(xlabel='x')
ax.tick_params(axis="y",colors="g")
ax.set_ylabel("y=$e^x$",color="g")

ax2 = ax.twinx()
ax2.plot(x,lny,"--",color=[.2,.3,.45],label="ln(x)")
ax2.tick_params(axis="y",colors=[.2,.3,.45])
ax2.set_ylabel("y=ln(x)",color=[.2,.3,.45])

# get both legends (I got help in here)
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2)

plt.tight_layout()
plt.show()
```
