# **DDC-12: Numpy correlation and matplotlib**
Use numpy to generate two correlated variables with a sample size of 101. Variable x is linearly spaced numbers between -.5 and +.5, and variable y is x plus uniformly distributed noise in the range of [-.5,.5]. Calculate the correlation coefficient using numpy, and use matplotlib to show the scatter plot with the correlation in the title, as in the figure below.

  <img src="images/img_ddc_012.png" alt="012" width="700">
</p>

```python
# Libs

import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
#?np.random.uniform
```

```python
x = np.linspace(-5,5,101)
y = x + np.random.uniform(-5,5,101)

corr = np.corrcoef(x,y)[1,0]
```

```python
plt.figure(figsize=(9,5))

plt.plot(x,y,"p",markerfacecolor=[.3,.7,.4,.8],markersize=9)
plt.gca().set(xlabel="x",ylabel="y",title=f"Correlation: {corr:.2f}")

plt.tight_layout()
plt.show()
```
