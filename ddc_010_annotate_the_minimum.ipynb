Write code to reproduce the figure below. Use the annotation() method on a matplotlib axis to show the text and wedge. The formula that defines the values is printed on the y-axis.

```python
# Libs

import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
x = np.linspace(-5,5,50)
y = x**2 + 10*np.cos(x)
min_point = np.argmin(y)
```

```python
# ?plt.Annotation
```

```python
_,ax = plt.subplots(1,figsize=(8,4))

ax.plot(x,y,"sg-")
ax.annotate(text=f'min: ({x[min_point]:.2f},{y[min_point]:.2f})',xy=(x[min_point],y[min_point]),
               arrowprops = dict(color=[.2,.4,.6],arrowstyle='fancy'),
               xycoords = 'data',
               xytext = (x[int(len(x)/2)],np.max(y)*.8),horizontalalignment='center',fontsize = 11)
ax.set(xlabel="x")
ax.set_ylabel("$f(x)=x^2+10cos(x)$",fontsize=13)


plt.tight_layout()
plt.show()
```
