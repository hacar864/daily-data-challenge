# **DDC-11: Subplots with gridspec**
Use the gridspec module in matplotlib to create a figure like the one below. The data are all random numbers; the main focus here is setting up the axes.

```python
# Libs

import numpy as np
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
rand_img = np.random.randn(100,51)

x = np.linspace(0,5,10)

a = np.random.normal(size=10)
b = np.random.randn(10)
c = np.random.random(10)
d = np.random.uniform(size=10)

k = np.random.normal(size=10)
l = np.random.randn(10)
m = np.random.random(10)
n = np.random.uniform(size=10)



print(rand_img.shape,x.shape,a.shape,b.shape,c.shape,d.shape)
```

```python
fig = plt.figure(figsize=(8,5))

gs = GridSpec(2,2)
ax0 = fig.add_subplot(gs[:,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,1])

ax0.imshow(rand_img)

ax1.plot(x,a,
         x,b,
         x,c,
         x,d)

ax2.plot(x,k,
         x,l,
         x,m,
         x,n)

plt.tight_layout()
plt.show()
```
