This is similar to DDC-12, but using pandas and seaborn.

Use numpy to generate two correlated variables with a sample size of 101. Variable x is linearly spaced numbers between -.5 and +.5, and variable y is x plus uniformly distributed noise in the range of [-.5,.5].

Import the data into a pandas dataframe, and use sns.regplot to make a visualization like this:
<p align="center">
  <img src="images/img_ddc_013.png" alt="013" width="700">
</p>

```python
# Libs

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")
```

```python
x       = np.linspace(-.5,.5,101)
y       = x + np.random.uniform(-.5,.5,101)
corr    = np.corrcoef(x,y)[0,1]

df = pd.DataFrame({"x":x,"y":y})
```

```python
plt.figure(figsize=(8,6))

h = sns.regplot(data=df,x='x',y='y',color=[.3,.7,.4,.5])
h.set(title=f'Correlation = {corr:.2f}')

plt.tight_layout()
plt.show()
```
