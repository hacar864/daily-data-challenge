```python
# Libs
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
# ?np.arange
```

```python
numbers = np.arange(-20,21,.5)
numbers
```

```python
def y1(x):
    return 5+3*(x**2)

def y2(x):
    return 3-2*x
```

```python
plt.figure(figsize=(8,5))

plt.plot(numbers,y1(numbers),numbers,y2(numbers))
plt.legend(["$y_1$","$y_2$"])   # :O
plt.gca().set(xlabel="Numbers",ylabel="Values")

plt.tight_layout()
plt.show()
```
